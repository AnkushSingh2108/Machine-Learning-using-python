```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.arange(20)
x
```

```python
theta = np.array([2,3])
noise = 5*np.random.randn(20)
```

```python
y_ideal = theta[1]*x + theta[0]
y_real = theta[1]*x + noise
```

```python
plt.plot(x,y_ideal,color = "orange", label = "Ideal Case")
plt.scatter(x,y_real,label = "Real Case")
plt.scatter(x,noise, label = "Noise Case")
plt.legend()
plt.show()
```
