```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
%matplotlib notebook
```

```python
X,y = make_regression(n_samples=1000, n_features=2,n_informative=2,noise=10)
X.shape,y.shape
```

```python
plt.subplot(1,2,1)
plt.scatter(X[:,0],y)

plt.subplot(1,2,2)
plt.scatter(X[:,1],y)
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure(figsize = (10,7))
ax = plt.axes(projection = '3d')

# Creating Plot
ax.scatter3D(X[:,0], X[:,1], y, color = "blue")
plt.title("3D Scatter Plot")
plt.show()
```

# Implementing Linear Regression ising sklearn

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.coef_ # this gives the coefficient "ax+by+c=0" 
```

```python
model.intercept_ # this gives theta0 or the value of c in "ax+by+c=0"
```

```python
model.predict([X[0],X[1]])  #this gives the predictions i.e. y values for X[0] abd X[1]
```

```python
y[0],y[1]
```

```python
model.score(X,y)
```
