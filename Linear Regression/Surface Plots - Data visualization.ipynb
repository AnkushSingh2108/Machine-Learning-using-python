```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib notebook
```

```python
X = pd.read_csv('./Linear Regression/Training Data/Linear_X_Train.csv')
Y = pd.read_csv('./Linear Regression/Training Data/Linear_Y_Train.csv')
```

```python
x = X.values
y = Y.values
```

```python
x,y = np.meshgrid(x,y)
```

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.plot_surface(x,y,x**2+y**2,cmap = 'rainbow')   # the parameters are : x_value,y_value,z_eqn
plt.title("Surface plot")
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.contour(x,y,x**2+y**2,cmap = 'rainbow')   # the parameters are : x_value,y_value,z_eqn
plt.title("Contour plot")
plt.show()
```
