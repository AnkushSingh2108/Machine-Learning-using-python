```python
import matplotlib.pyplot as plt
x1 = [5,6,7,89,5466]
y1 = [1,2,5,6,845]
plt.plot(x1,y1, label="Line 1")

x2 = [5,6,7,89,54]
y2 = [1,2,5,6,8555]
plt.plot(x2,y2, label="Line 2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multile line plot!!")
plt.legend()
plt.show()
```

```python
x = [1,2,45,54,85,96,65,23]
y = [23,65,96,85,54,45,2,1]
plt.plot(x,y, label="Line 2", color = "black", linestyle = ":", linewidth = 5, marker = "*", markersize = 10,
         markerfacecolor = "red")
plt.xlabel("X-axis",color = "blue")
plt.ylabel("Y-axis", color = "blue")
plt.title("Multiple line plot!!",color = "blue")
plt.legend()
plt.show()
```
