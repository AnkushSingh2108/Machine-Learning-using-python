# Univariate Selection

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

```python
data = pd.read_csv("./Mobile Price Classicication/mobile price pred train.csv")
```

```python
data.head()
```

```python
x = data.iloc[:,0:20]
y = data.iloc[:,-1]
```

# 1) Apply SelectKBest Method of Feature Selection
- Top 10 Features

```python
best_feature = SelectKBest(score_func= chi2, k = 10)
```

```python
fit = best_feature.fit(x,y)
```

```python
fit.scores_
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x.columns)
```

```python
feature_scores = pd.concat([dfcolumns, dfscores], axis =1)
```

```python
feature_scores.columns = ["Features", "Score"]
```

```python
feature_scores
```

```python
feature_scores.sort_values(by = 'Score', ascending = False)
```

```python
ten_features = feature_scores.sort_values(by = 'Score', ascending = False).head(10)['Features'].values
```

# 2) Feature Importance Method of Feature Selection
- Its a Decision Tree based Model/Class from scikit learn
- Their is a property called importance so that it can tell the importance of each feature.
- The more important the feature is the higher value it will have

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Since RandomForestClassifier is a class we can initialize and create an object
model = RandomForestClassifier()
model.fit(x,y)
```

```python
model.feature_importances_
```

```python
feature_importance = pd.DataFrame(model.feature_importances_, index = x.columns, columns = ['Importance'])
feature_importance = feature_importance.sort_values(by = 'Importance', ascending = False)
```

```python
plt.figure(figsize = (20,10))
plt.bar(feature_importance.index, feature_importance['Importance'])
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
rfc = RandomForestClassifier(max_depth=10)
scores = cross_val_score(rfc,x,y,cv = 10)  # cv is the number of folds of training and testing you want to do
```
