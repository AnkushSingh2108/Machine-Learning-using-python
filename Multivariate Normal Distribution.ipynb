```python
import numpy as np
import matplotlib.pyplot as plt
```

# Multivariate Normal Distribution

```python
#Bivariate

mean = np.array([0.0,0.0])
covariance = np.array([[10,2],[20,2]])

mean2 = np.array([1.0,2.0])
covariance2 = np.array([[5.3,2.2],[3.2,2.1]])

dist = np.random.multivariate_normal(mean,covariance,500)
dist2 = np.random.multivariate_normal(mean2,covariance2,500)
print(dist.shape)
print(dist.shape)

plt.scatter(dist[:,0], dist[:,1])
plt.scatter(dist2[:,0], dist2[:,1])
plt.show()
```
