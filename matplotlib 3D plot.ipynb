```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  #this module is used for 3D plotting
import numpy as np
%matplotlib notebook
#because of this only in jupyter notebook we can rotate the 3D plot
```

```python
# print(dir(np))
```

```python
X = np.arange(-5,5,0.25)
Y = np.arange(-5,5,0.25)
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X**2 + Y**2)
S = np.sqrt(X**5 + Y**5)
Z = np.exp(R)
A = np.exp(-R)
D = np.exp(S)
F = np.exp(-S)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X,Y,Z, rstride = 1, cstride = 2)
plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X,Y,A, rstride = 1, cstride = 2)
plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X,Y,D, rstride = 1, cstride = 2)
plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X,Y,F, rstride = 1, cstride = 2)
plt.show()
```
